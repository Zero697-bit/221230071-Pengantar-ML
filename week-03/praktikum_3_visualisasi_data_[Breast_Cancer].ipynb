Tahap 1 Persiapan Awal Install Package yang di butuhkan

```python
# Import library yang dibutuhkan
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_breast_cancer
```

Tahap 2a Load Dataset

```python
# Tahap 2a: Load Dataset Breast Cancer
data = load_breast_cancer()

# Membentuk DataFrame
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target  # 0 = malignant (ganas), 1 = benign (jinak)

# Simpan ke file CSV
df.to_csv("breast_cancer.csv", index=False)
print("Dataset disimpan sebagai breast_cancer.csv")
```

TAHAP 2B Eksplorasi Struktur Data

```python
# Tahap 2b: Eksplorasi Struktur Data
print("=== BREAST CANCER DATASET ===")
print(f"Shape: {df.shape}")
print(f"Columns: {df.columns.tolist()}")
print(f"Data Types:\n{df.dtypes}")
print(f"\nMissing Values:\n{df.isnull().sum()}")
print(f"\nBasic Statistics:")
print(df.describe())
```

Menampilkan Dataset

```python
# Menampilkan dataset
df
```

Analisis Fitur yang Membedakan Malignant dan Benign

```python
# Tahap 4: Analisis Fitur yang Membedakan Malignant dan Benign
# Mengelompokkan data berdasarkan target
group_mean = df.groupby('target').mean()

# Menghitung selisih absolut rata-rata antara dua kelas
mean_diff = abs(group_mean.loc[0] - group_mean.loc[1])

# Mengambil 10 fitur dengan perbedaan paling besar
top_diff = mean_diff.sort_values(ascending=False).head(10)

# Visualisasi
plt.figure(figsize=(8,6))
sns.barplot(x=top_diff.values, y=top_diff.index, palette="Set2")
plt.title("Top 10 Fitur yang Paling Membedakan Malignant dan Benign")
plt.xlabel("Perbedaan Rata-rata (|Mean Difference|)")
plt.ylabel("Fitur")
plt.show()

top_diff
```

Analisis Korelasi antar Fitur

```python
# Tahap 5: Analisis Korelasi antar Fitur
corr_matrix = df.corr()

# Melihat korelasi antara semua fitur terhadap target
corr_target = corr_matrix["target"].sort_values(ascending=False)
print(corr_target)
```

Heatmap Korelasi Antar Fitur Tumor

```python
plt.figure(figsize=(14,10))
sns.heatmap(corr_matrix, cmap='coolwarm', linewidths=0.5)
plt.title("Heatmap Korelasi Antar Fitur Tumor")
plt.show()
```

Analisis Subgroup pada Kelas Malignant & Benign dan Visualisasi Subgroup Tumor Menggunakan PCA

```python
# Tahap: Analisis Subgroup pada Kelas Malignant & Benign

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# === 1. Standarisasi fitur (agar semua fitur punya skala yang sama) ===
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df.drop('target', axis=1))

# === 2. Reduksi dimensi menggunakan PCA (2 dimensi untuk visualisasi) ===
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_data)

# Simpan hasil PCA ke dataframe
df['PCA1'] = pca_result[:, 0]
df['PCA2'] = pca_result[:, 1]

# === 3. Visualisasi hasil PCA ===
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=df, x='PCA1', y='PCA2', hue='target',
    palette={0: 'red', 1: 'green'}, alpha=0.7
)
plt.title('Visualisasi Subgroup Tumor Menggunakan PCA\n(0 = Malignant, 1 = Benign)')
plt.xlabel('PCA Komponen 1')
plt.ylabel('PCA Komponen 2')
plt.show()
```

Distribusi Fitur Kunci Berdasarkan Kelas Tumor

```python
# Pilih beberapa fitur penting untuk dianalisis
key_features = ['mean radius', 'mean texture', 'mean perimeter', 'mean area', 'mean smoothness']

plt.figure(figsize=(12, 8))
for i, feature in enumerate(key_features, 1):
    plt.subplot(2, 3, i)
    sns.violinplot(x='target', y=feature, data=df, palette=['red', 'green'])
    plt.title(f'Distribusi {feature}')
    plt.xlabel('Target (0 = Malignant, 1 = Benign)')
    plt.ylabel('Nilai')
plt.tight_layout()
plt.show()
```
